## Extract

```python
import pandas as pd
import numpy as np
```

```python
df_arquivo = pd.read_excel("Desktop\Exercício DIO.xlsx")
```

```python
print(df_arquivo)
```

## Transform

```python
exercicios_mais_executados = df_arquivo['Exercicio'].value_counts()
print(exercicios_mais_executados)
```

## Load

```python
exercicios_mais_executados.to_csv("Exercicios mais executados.csv")
```

```python
import matplotlib.pyplot as plt
```

```python
exercicios_mais_executados.plot(kind='bar')
plt.xlabel('Exercicio')
plt.ylabel('Dias')
plt.title('Frequência de treinos')
plt.show()
```
